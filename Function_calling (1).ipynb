```python
def get_current_Datetime(timezone:str="UTC") -> dict:
    """Fetches the curent date and time
    taking timezone as input
    string data type
    returns a dict containing the curr date and time of the specified timezone
    or an error msg if the timezone is invalid
    """
    print(f"function calling for checking the whether the function is invoked or not {timezone}")
    try:
        tz=pytz.timezone(timezone)
        now=datetime.datetime.now(tz)
        return {
            "timezone":timezone,
            "current_datetime":now.strftime("%Y-%m-%d %H:%M:%S")
        }
    except pytz.exceptions.UnknownTimeZoneError as e:
        return {"error":str(e)}

get_current_Datetime("Asia/Kolkata")
```

```python
from google.genai import types
response=client.models.generate_content(
    model='gemini-2.5-flash',
    contents="Current Data and time in India",
    config=types.GenerateContentConfig(
        tools=[get_current_Datetime],
        thinking_config=types.ThinkingConfig(
            include_thoughts=True,
        )
    )
)

print(response.automatic_function_calling_history[2])
```
