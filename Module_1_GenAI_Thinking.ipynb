# GEMINI JSON AND ENUM STRUCTURED OUTPUT

```python
from google import genai
from pydantic import BaseModel, Field
from typing import List
from google.genai import types # Import types


class Programming_Language(BaseModel):
    language_name: str
    Year_created: str
    creator: str
    use_case: List[str]

prompt_text = """List Popular Programming Language and name its publishing year, creator, two use cases """

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt_text,
    config={
        "response_mime_type": "application/json",
        "response_schema": list[Programming_Language],
    },
)
# Use the response as a JSON string.
print(response.text)
```

```python
import pandas as pd
import json

# Assuming 'response.text' contains the JSON string from the previous cell
json_data = json.loads(response.text)

df = pd.DataFrame(json_data)

display(df)
```

```python
class Railway(BaseModel):
    sl_no:int
    station:str
    zone:str
    code:str
    division:str
    earning:int

prompt_text = """extract all the information as per the given json structure only """

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=[image,prompt_text],
    config={
        "response_mime_type": "application/json",
        "response_schema": list[Railway],
    },
)
# Use the response as a JSON string.
print(response.text)

import pandas as pd
import json

# Assuming 'response.text' contains the JSON string from the previous cell
json_data = json.loads(response.text)

df = pd.DataFrame(json_data)

display(df)
```

ENUM

```python
from google import genai

import enum
from pydantic import BaseModel

class Grade(enum.Enum):
    A_PLUS = "a+"
    A = "a"
    B = "b"
    C = "c"
    D = "d"
    F = "f"

class Recipe(BaseModel):
  recipe_name: str
  rating: Grade

response = client.models.generate_content(
    model='gemini-2.5-flash',
    contents='List 10 home-baked cookie recipes and give them grades based on tastiness.',
    config={
        'response_mime_type': 'text/x.enum',
        'response_schema': list[Recipe],
    },
)

print(response.text)
```
