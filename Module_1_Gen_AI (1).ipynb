```python
# -------------------------------
# System Instructions
# -------------------------------
system_instructions = """
You are a corporate legal assistant.
Your job is to extract the following from any B2B contract text:
1. Contract start and end dates,
2. Payment terms,
3. Termination conditions,
4. Any penalty or liability clauses.
Output only what you find, in tabular form.
"""

# -------------------------------
# Contract Content
# -------------------------------
contents = """
This agreement is entered into on May 1st, 2024 and shall remain in effect until April 30, 2025.
Payment is due within 45 days of invoice date. Late payments may incur a penalty of 3% monthly.
Either party may terminate this agreement with 60 days' notice.
"""

# -------------------------------
# Model Call
# -------------------------------
response = client.models.generate_content(
    model="gemini-2.5-flash",  # or "gpt-5", "gemini-1.5-flash", etc.
    config=types.GenerateContentConfig(
        system_instruction=system_instructions
    ),
    contents=contents
)

# -------------------------------
# Output
# -------------------------------
print(response.text)
```

```python
import os
from google import genai
from google.genai import types
from dotenv import load_dotenv
load_dotenv()
client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))
client
```

```python
content = """
In a cheerful female voice: Hey, how are you today?
In a deep male voice: I'm doing great, thanks for asking!
In a soft female voice: The weather is really nice, isn't it?
In a young male voice: Absolutely, perfect for a walk in the park!
In an excited female voice: Did you see the festival decorations?
In a calm male voice: Yes, they look beautiful and vibrant.
In a playful female voice: Let's go try some of the street food.
In a serious male voice: Good idea, we should also stay safe.
In a happy female voice: I love this city during festive times!
In a warm male voice: Me too, it feels like everyone is smiling.
"""
response = client.models.generate_content(
    model="gemini-2.5-flash-preview-tts",
    contents=content,
    config=types.GenerateContentConfig(
        response_modalities=["AUDIO"],
        speech_config = types.SpeechConfig(
            voice_config=types.VoiceConfig(
                prebuilt_voice_config=types.PrebuiltVoiceConfig(
                    voice_name= 'Kore'
                )
            )
        ),
    )
)
```

```python
data=response.candidates[0].content.parts[0].inline_data.data
```

```python
# .wav
# channel --> 1(mono),2{stereo}
# samplewidth -->  clearity 8,16 bits
# framerate -->
import wave
```

```python
def wave_file(filename,input):
    with wave.open(filename,'wb') as wf:
        wf.setnchannels(1)
        wf.setsampwidth(2)
        wf.setframerate(24000)
        wf.writeframes(input)
```

```python
file_name='output1.wav'
wave_file(file_name,data)
from IPython.display import Audio

# Play the generated audio file
Audio("output1.wav")
```
